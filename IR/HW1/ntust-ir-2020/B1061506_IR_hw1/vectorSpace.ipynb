```python
voc = set()
for q in queries:
    adr = 'queries/' + q + '.txt'
    words = open(adr,mode = 'r').read()
    words = words.split(" ")
    for word in words:
        voc.add(word)
```

```python
voc = list(voc)
```

```python
vector_Q = [0]*len(voc)
vector_D = [0]*len(voc)
```

```python

history = pd.DataFrame(columns = ['word','idf'])
tf = pd.DataFrame(columns= ['doc','tf'])
for q in queries:
    adr = 'queries/' + q +'.txt'
    query = open(adr,mode = 'r').read()
    query = query.split(" ")
    df = pd.DataFrame(columns = ['doc','score'])
    vector_Q = [0]*len(voc)
    for w in query:
        if w in voc:
            if history[history.word == w].size == 0:
                idf = 0
                for d in docs:
                    if (d != ''):
                        adr = 'docs/' + d + '.txt'
                        words = open(adr,mode = 'r').read()
                        words = words.split(" ")
                        if w in words:
                            idf += 1
                dict_w = {
                    'word': w,
                    'idf': idf
                }
                history = history.append(dict_w,ignore_index = True)
            else:
                idf = history[history.word == w]['idf'].values[0]
            vector_Q[voc.index(w)] +=math.log(1 + 4191/(idf),10) 
    counter = 0
    for doc in docs:
        adr = 'docs/' + doc + '.txt'
        words = open(adr,mode = 'r').read()
        words = words.split(" ")
        vector_D = [0]*len(voc)
        if tf[tf.doc == doc].size == 0:
            for w in words:
                vector_D[voc.index(w)] += 1
                if w in voc:
                    vector_D[voc.index(w)] += 1
                v = {
                    'doc': doc,
                    'tf': vector_D
                    }
            tf = tf.append(v, ignore_index=True)
        else:
            vector_D = tf[tf.doc == doc]['tf'].values[0]
        
        dot = sum(a*b for a, b in zip(vector_Q, vector_D))
        norm_a = sum(a*a for a in vector_Q) ** 0.5
        norm_b = sum(b*b for b in vector_D) ** 0.5

    # Cosine similarity
        if (norm_a*norm_b) == 0:
            cos_sim = 0
        else:
            cos_sim = dot / (norm_a*norm_b)
        counter += 1
        #print(counter, cos_sim)
        dic = {
            'doc': doc,
            'score': cos_sim
        }
        df = df.append(dic,ignore_index=True)
    
    sortdf = df.sort_index(by = 'score',ascending = False)
    fo.write(q + ',')
    for i in range (4191):
        fo.write(str(sortdf.iloc[i,0]) + ' ')
    fo.write('\n')
    print (q)
    sortdf.head()
```
