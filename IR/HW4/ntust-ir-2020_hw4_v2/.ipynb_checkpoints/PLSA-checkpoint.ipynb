```python
import pandas as pd
import numpy as np
import math
import random
```

```python
doc_list = open('doc_list.txt',mode = 'r').read()
query_list = open('query_list.txt',mode = 'r').read()
doc_list = doc_list.split(sep = '\n')
query_list = query_list.split(sep = '\n')
```

```python
docs = []
for doc in doc_list:
    adr = 'docs/' + doc + '.txt'
    words = open(adr,mode = 'r').read()
    words = words.split(" ")
    docs.append(words)
```

```python
querys = []
for query in query_list:
    adr = 'queries/' + query + '.txt'
    words = open(adr,mode = 'r').read()
    words = words.split(" ")
    querys.append(words)
```

```python
q_word = {}
for query in querys:
    for word in query:
        q_word[word] = q_word.get(word, 0) + 1
    
```

```python
#算tf
tf = []
df = {}
idf = {}
for doc in docs:
    tmp = {}
    for word in doc:
        tmp[word] = tmp.get(word, 0) + 1
    tf.append(tmp)
    for k in tmp.keys():
        df[k] = df.get(k,0) + 1
#算idf

for k, v in df.items():
    idf[k] = math.log(  ( len(docs) - v + 0.5 )) / math.log((v + 0.5))
    
```

```python
N = 14955
M = 226
K = 32
lamda = np.random.random([N,K])
theta = np.random.random([K,M])
for i in range(0, N):
    normalization = sum(lamda[i, :])
    for j in range(0, K):
        lamda[i, j] /= normalization

for i in range(0, K):
    normalization = sum(theta[i, :])
    for j in range(0, M):
        theta[i, j] /= normalization
```

```python
tkWi = []
tkDj = lamda

for i in range(K):
    pT = {}
    it = 0
    for k in q_word.keys():
        pT[k] = theta[i,it]
        it += 1
    tkWi.append(pT)
    
```

```python
twd = []
for i in range(N):
    tmp = {}
    for j in tkWi[0].keys():
        tmp[j] = np.zeros([K])
    twd.append(tmp)
```

```python
lL = -100000000
for bech in range(150):
    print('bach: ',bech)
    #E-step
    for i in range(N):
        for j in tkWi[0].keys():
            denominator = 0.0
            for k in range(K):
                twd[i][j][k] = tkWi[k][j] * tkDj[i][k]
                denominator += twd[i][j][k]
            if denominator == 0:
                for k in range(K):
                    twd[i][j][k] = 0.0
            else:
                for k in range(K):
                    twd[i][j][k] /= denominator

    #M-step
    # update tkWi
    for k in range(K):
        print('up tkWi: ',k)
        denominator = 0.0
        for j in tkWi[0].keys():
            tkWi[k][j] = 0.0
            for i in range(14955):
                tkWi[k][j] += tf[i].get(j,0) *  twd[i][j][k]

            denominator += tkWi[k][j]
        if denominator == 0:
            for j in tkWi[0].keys():
                tkWi[k][j] = 1.0 / M
        else:
            for j in tkWi[0].keys():
                tkWi[k][j] /= denominator
    # update tkDj
    for i in range(14955):
        for k in range(K):
            tkDj[i][k] = 0.0
            denominator = 0.0
            for j in tkWi[0].keys():
                tkDj[i][k] += tf[i].get(j,0) * twd[i][j][k]
                denominator += tf[i].get(j,0)
            if denominator == 0:
                tkDj[i][k] = 1.0 / K
            else:
                tkDj[i][k] /= denominator

    #LOG
    L = 0.0
    for i in range(N):
        for j in tkWi[0].keys():
            tmp = 0.0
            for k in range(K):
                tmp += tkWi[k][j] * tkDj[i][k]
            if tmp > 0:
                L += tf[i].get(j,0) * math.log(tmp)
            
    print(L)
    if lL != 1 and (L - lL) < 10.0:
        break
    lL = L
    
```

```python
his = {}


alpha = 0.7
bata = 0.2
all_rank = []

for query in querys:
    rank = []
    print(len(all_rank))    
    for i in range(14955):
        p = 1
        for word in query:
            pWiDj = tf[i].get(word,0) / len(docs[i])
            sigmaK = 0
            for tk in range(K):
                sigmaK += tkWi[tk][word] * tkDj[i][tk]

            pWiBG = his.get(word,0)
            if pWiBG == 0:
                uP = 0
                dP = 0
                for itr,doc in enumerate(docs):
                    uP += tf[itr].get(word,0)
                    dP += len(doc)
                
                pWiBG = uP / dP
                his[word] = pWiBG

            tmp = alpha * pWiDj + bata * sigmaK + (1 - alpha - bata) * pWiBG 
            
            p*=tmp 
        rank.append(p)
    all_rank.append(rank)
```

```python
fo = open('output.txt','w')
fo.write('Query,RetrievedDocuments\n')

for i in range(100):
    
    list1 = all_rank[i]
    list2 = doc_list      
    list1, list2 = zip(*sorted(zip(list1, list2)))

    fo.write(query_list[i] + ',')
    for x in range (N):
        fo.write(str(list2[14954 - x]) + ' ')
    fo.write('\n')
```
